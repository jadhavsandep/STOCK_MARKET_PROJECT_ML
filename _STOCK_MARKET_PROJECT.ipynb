```python
# Importing necessary libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np   # For numerical computations
import matplotlib.pyplot as plt  # For basic plotting
import seaborn as sns  # For advanced statistical visualizations
import plotly.express as px  # For interactive visualizations

# Display plots inline
%matplotlib inline
```

```python
# Load data from the specified directory
data1 = pd.read_csv(r"C:\Users\SANDEEP JADHAV\OneDrive\Desktop\stockp-market\bond_etfs.csv")
data2 = pd.read_csv(r"C:\Users\SANDEEP JADHAV\OneDrive\Desktop\stockp-market\equity_etfs.csv")
data3 = pd.read_csv(r"C:\Users\SANDEEP JADHAV\OneDrive\Desktop\stockp-market\mag7.csv")
data4 = pd.read_csv(r"C:\Users\SANDEEP JADHAV\OneDrive\Desktop\stockp-market\metals_etfs.csv")
```

```python
# Concatenate datasets side-by-side (ensure they have the same number of rows)
Data = pd.concat([data1, data2, data3, data4], axis=1)

# Display the combined dataset
Data.head()
```

```python
# Histogram for a single column
plt.figure(figsize=(8, 6))
Data.iloc[:, 1].hist(bins=20)
plt.title("Distribution of Column 1")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot for a specific set of columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=Data.iloc[:, 1:10])
plt.title("Box Plot of Columns 1 to 10")
plt.show()
```

```python
# Scatter plot for two columns
plt.figure(figsize=(8, 6))
plt.scatter(Data.iloc[:, 1], Data.iloc[:, 2])
plt.xlabel("Column 1")
plt.ylabel("Column 47")
plt.title("Scatter Plot Between Column 1 and Column 47")
plt.show()
```

```python
# Pair plot for numerical columns
sns.pairplot(Data.iloc[:, 1:5])
plt.show()
```

```python
# Bar plot for specific columns
Data_sample = Data.iloc[:10, 1:3]  # First 10 rows for demonstration
Data_sample.plot(kind="bar", figsize=(12, 6))
plt.title("Bar Plot of Columns 1 and 2")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select numerical columns
numerical_columns = Data.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(Data[column].dropna(), bins=20, alpha=0.7, edgecolor='green')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
